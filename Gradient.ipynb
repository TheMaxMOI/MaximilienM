```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

def gradient(height, width, f):
    return np.array([[f(i, j) for j in range(width)] for i in range(height)])

def f(n,h=None,w=None):
    if n==1:
        return lambda x,y: atan(y/x)*2/pi if x>0 else 0
    elif n==2:
        return lambda x,y: 1/(1+exp(-y/x)) if x>0 else 0
    elif w is not None:
        if n==3:
            return lambda x,y: atan(w/pi/y)*2/pi if y>0 else 1
        elif n==4:
            return lambda x,y: 1-x/w
    elif w is not None and h is not None:
        if n==5:
            return lambda x,y: -(y-x)/w+h
    elif n==6:
        return lambda x,y: sin(x)*2/pi
    elif n==7:
        return lambda x,y: f(6)(y,x) #sin(y)*2/pi
    elif n==8:
        return lambda x,y: f(6)(x*y,0) #sin(y*x)*2/pi
    elif n==9:
        return lambda x,y: (cos(x)+cos(y))/2
    elif n==10:
        return lambda x,y: f(9)(x,y+pi) #(cos(x)-cos(y))/2
    elif n==11:
        return lambda x,y: (cos(x)*cos(y))
    elif n==12:
        return lambda x,y: int(f(11)(x,y)+1)

def visualisation(gradient):
    h, w = np.shape(gradient)
    img = np.empty((h, w))
    for i in range(h):
        for j in range(w):
            img[i, j] = int(255 * gradient[i, j])
    plt.matshow(img)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

visualisation(gradient(100,300,f(8)))
```
